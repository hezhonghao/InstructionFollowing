```python
# Loading model
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from datasets import load_dataset
import math

# Load the model and tokenizer
model_name = "gpt2"  # You can change this to any model you want to evaluate
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()

# Load Wikitext-103 dataset
dataset = load_dataset("wikitext", "wikitext-103-v1", split="test")
```

```python
# A demonstration of language modeling tasks on wikitext-103 # TODO
```

```python

def calculate_perplexity(model, tokenizer, text):
    encodings = tokenizer(text, return_tensors="pt")
    max_length = model.config.max_position_embeddings
    stride = 512
    seq_len = encodings.input_ids.size(1)

    nlls = []
    prev_end_loc = 0
    for begin_loc in range(0, seq_len, stride):
        end_loc = min(begin_loc + max_length, seq_len)
        trg_len = end_loc - prev_end_loc  # may be different from stride on last loop
        input_ids = encodings.input_ids[:, begin_loc:end_loc]
        target_ids = input_ids.clone()
        target_ids[:, :-trg_len] = -100

        with torch.no_grad():
            outputs = model(input_ids, labels=target_ids)
            neg_log_likelihood = outputs.loss

        nlls.append(neg_log_likelihood)

        prev_end_loc = end_loc
        if end_loc == seq_len:
            break

    ppl = torch.exp(torch.stack(nlls).mean())
    return ppl.item()

# Evaluate on a subset of Wikitext-103
num_samples = 1000  # Adjust this based on your computational resources
total_ppl = 0

for i in range(num_samples):
    text = dataset[i]["text"]
    if text.strip():  # Skip empty lines
        ppl = calculate_perplexity(model, tokenizer, text)
        total_ppl += ppl
        print(f"Sample {i+1}/{num_samples}, Perplexity: {ppl:.2f}")

average_ppl = total_ppl / num_samples
print(f"\nAverage Perplexity on Wikitext-103 (subset): {average_ppl:.2f}")
```

## Fine-tune on Wikitext-103

```python
(claude code)
import torch
from transformers import GPT2LMHeadModel, GPT2TokenizerFast, TextDataset, DataCollatorForLanguageModeling
from transformers import Trainer, TrainingArguments
from datasets import load_dataset

# Step 1: Load the model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2TokenizerFast.from_pretrained(model_name)

# Set padding token  #NEP what is this?
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = model.config.eos_token_id

# Step 2: Prepare the dataset (subset)
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, max_length=512)

dataset = load_dataset("wikitext", "wikitext-103-v1")
train_subset = dataset["train"].select(range(10000))  # Use only first 10,000 examples
tokenized_dataset = train_subset.map(tokenize_function, batched=True, remove_columns=["text"])


# Step 3: Create data collator
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)


# Step 4: Define training arguments (modified)
training_args = TrainingArguments(
    output_dir="./gpt2-quick-finetuned-wikitext103",
    overwrite_output_dir=True,
    num_train_epochs=1,  # Keep this at 1 for quick training
    per_device_train_batch_size=8,  # Increase if you have more GPU memory
    save_steps=1000,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=100,
    learning_rate=5e-4,  # Slightly higher learning rate
    warmup_steps=100,
    weight_decay=0.01,  # Add some regularization
    fp16=True if torch.cuda.is_available() else False,  # Use mixed precision training if your GPU supports it
)

# Step 5: Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_dataset,
)



# Step 6: Fine-tune the model
trainer.train()

# Step 7: Save the fine-tuned model
model.save_pretrained("./gpt2-finetuned-wikitext103")
tokenizer.save_pretrained("./gpt2-finetuned-wikitext103")
```

```python
# Step 8 evaluation of fine-tuned model  (claude code)
from tqdm import tqdm

def calculate_perplexity(model, tokenizer, dataset, max_length=512, batch_size=1):
    model.eval()
    total_loss = 0
    total_length = 0

    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model.to(device)

    with torch.no_grad():
        for i in tqdm(range(0, len(dataset), batch_size), desc="Calculating Perplexity"):
            batch = dataset[i:i+batch_size]

            # Check if the batch has any content
            if not batch['text']:
                continue

            encodings = tokenizer(batch["text"], truncation=True, max_length=max_length, return_tensors="pt", padding=True)
            input_ids = encodings.input_ids.to(device)

            # Check if input_ids is not empty
            if input_ids.numel() == 0:
                continue

            outputs = model(input_ids, labels=input_ids)

            total_loss += outputs.loss.item() * input_ids.size(1)
            total_length += input_ids.size(1)

            # Clear GPU cache
            if device == 'cuda':
                torch.cuda.empty_cache()

    if total_length == 0:
        raise ValueError("No valid inputs were processed. Check your dataset and tokenization.")

    avg_loss = total_loss / total_length
    perplexity = torch.exp(torch.tensor(avg_loss))
    return perplexity.item()

# Usage:
test_subset = dataset["test"].select(range(1000))  # Use only first 1,000 test examples
try:
    test_perplexity = calculate_perplexity(model, tokenizer, test_subset, batch_size=1)
    print(f"Perplexity on test subset: {test_perplexity:.2f}")
except ValueError as e:
    print(f"Error: {e}")
    print("Please check your dataset and ensure it contains valid text data.")
```
